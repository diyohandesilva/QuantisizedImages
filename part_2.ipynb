```
import numpy as np
from skimage import io
from sklearn.cluster import KMeans
from urllib.request import urlopen, Request
from PIL import Image


                         #place the link here
original = io.imread('https://i.imgur.com/EcCezG7.png')



#edit the k value here 
no_colors = 6

arr = original.reshape((-1, 3))
kmeans = KMeans(n_clusters= no_colors, random_state=42).fit(arr)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
less_colorsnew = centers[labels].reshape(original.shape).astype('uint8')


io.imshow(less_colorsnew)
```
